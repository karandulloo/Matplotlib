```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#loading the dataset and populating the first and last 5 records
prisoners=pd.read_csv("prisoners.csv")
prisoners.head(5)
```

```python
prisoners.tail(5)
```

```python
prisoners.info()
```

```python
#dropping zero values from the rows using loc mehtod
df = prisoners.loc[(prisoners != 0).all(axis=1), :]
df
```

# total of all rows but also taking the year column data which is not acceptable hence creating my own column list

```python
col_list= list(df)
col_list.remove('YEAR')
col_list
```

```python
df['total_benefitted'] = df[col_list].sum(axis=1)
df
```

```python
# plt.bar(
#     x=df['STATE/UT'],
#     height=df['total_benefitted'],
#     width=0.8,
#     align='center'
# )

df.plot(x='STATE/UT', y='total_benefitted', kind='bar') 
plt.show()
```

```python
plt.pie(
    df['total_benefitted'],
    labels=df['STATE/UT'],
    shadow=False,
    startangle=90,
    # with the percent listed as a fraction
    #autopct='%1.1f%%',
    )

# View the plot drop above
plt.axis('equal')

# View the plot
plt.show()
```
